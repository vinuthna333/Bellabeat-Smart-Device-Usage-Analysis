```python
# Calculate percentage of zero-step user-day records
zero_step_percentage = (zero_step_count / len(daily_activity)) * 100
zero_step_percentage
```

```python
# Count zero-step records per user
zero_steps_per_user = daily_activity[daily_activity["TotalSteps"] == 0]["Id"].value_counts()

zero_steps_per_user
```

```python
# Calculate total days per user
total_days_per_user = daily_activity["Id"].value_counts()

# Combine with zero-step counts
zero_analysis = pd.DataFrame({
    "TotalDays": total_days_per_user,
    "ZeroStepDays": zero_steps_per_user
}).fillna(0)

# Calculate percentage
zero_analysis["ZeroStepPercentage"] = (
    zero_analysis["ZeroStepDays"] / zero_analysis["TotalDays"]
) * 100

zero_analysis.sort_values("ZeroStepPercentage", ascending=False)
```

```python
# Check zero-step days by weekday
daily_activity[daily_activity["TotalSteps"] == 0]["DayOfWeek"].value_counts()
```

```python
# Correlation matrix
daily_activity[["TotalSteps", "Calories", "VeryActiveMinutes"]].corr()
```

```python
# Remove users with less than 30 total recorded days
reliable_users = zero_analysis[zero_analysis["TotalDays"] >= 30]

# Remove users with more than 40% zero-step days
reliable_users = reliable_users[reliable_users["ZeroStepPercentage"] <= 40]

# Keep only reliable users in dataset
daily_activity_cleaned = daily_activity[
    daily_activity["Id"].isin(reliable_users.index)
]
```

```python
daily_activity_cleaned = daily_activity_cleaned.copy()
```

```python
# Create activity level categories
def categorize_steps(steps):
    if steps < 5000:
        return "Sedentary"
    elif steps < 10000:
        return "Moderately Active"
    else:
        return "Active"

daily_activity_cleaned["ActivityLevel"] = (
    daily_activity_cleaned["TotalSteps"].apply(categorize_steps)
)
```

```python
daily_activity_cleaned["ActivityLevel"].value_counts()
```

```python
activity_percentage = (
    daily_activity_cleaned["ActivityLevel"].value_counts(normalize=True)
) * 100

activity_percentage
```

```python
# Calculate average steps per user
user_avg_steps = (
    daily_activity_cleaned
    .groupby("Id")["TotalSteps"]
    .mean()
)

# Categorize users by average steps
def categorize_avg_steps(steps):
    if steps < 5000:
        return "Sedentary User"
    elif steps < 10000:
        return "Moderately Active User"
    else:
        return "Highly Active User"

user_segments = user_avg_steps.apply(categorize_avg_steps)

user_segments.value_counts()
```

```python
# Check final dataset size after cleaning
print("Original number of records:", len(daily_activity))
print("Cleaned number of records:", len(daily_activity_cleaned))

# Calculate percentage of data retained
retained_percentage = (len(daily_activity_cleaned) / len(daily_activity)) * 100
print("Percentage of data retained:", round(retained_percentage, 2), "%")

# Check number of remaining users
print("Remaining unique users:", daily_activity_cleaned["Id"].nunique())
```

```python
# Calculate average calories per user
user_avg_calories = (
    daily_activity_cleaned
    .groupby("Id")["Calories"]
    .mean()
)
```

```python
# Combine user segment and average calories into one dataframe
user_summary = pd.DataFrame({
    "UserSegment": user_segments,
    "AvgCalories": user_avg_calories
})

# View first few rows
user_summary.head()
```

```python
# Calculate average calories for each user segment
segment_calorie_analysis = (
    user_summary
    .groupby("UserSegment")["AvgCalories"]
    .mean()
    .sort_values(ascending=False)
)

segment_calorie_analysis
```

```python
# We are checking how strongly intense activity (VeryActiveMinutes) is related to calorie expenditure
daily_activity_cleaned[["VeryActiveMinutes", "Calories"]].corr()
```

```python
# Normalize TotalSteps using Min-Max scaling
daily_activity_cleaned["Steps_Normalized"] = (
    (daily_activity_cleaned["TotalSteps"] - daily_activity_cleaned["TotalSteps"].min())
    /
    (daily_activity_cleaned["TotalSteps"].max() - daily_activity_cleaned["TotalSteps"].min())
)

# Normalize VeryActiveMinutes using Min-Max scaling
daily_activity_cleaned["VeryActive_Normalized"] = (
    (daily_activity_cleaned["VeryActiveMinutes"] - daily_activity_cleaned["VeryActiveMinutes"].min())
    /
    (daily_activity_cleaned["VeryActiveMinutes"].max() - daily_activity_cleaned["VeryActiveMinutes"].min())
)
```

```python
# Create composite activity score
# We give equal weight to Steps and Very Active Minutes
daily_activity_cleaned["ActivityScore"] = (
    0.5 * daily_activity_cleaned["Steps_Normalized"] +
    0.5 * daily_activity_cleaned["VeryActive_Normalized"]
)

# Check first few rows
daily_activity_cleaned[["TotalSteps", "VeryActiveMinutes",
                        "Steps_Normalized", "VeryActive_Normalized",
                        "ActivityScore"]].head()
```

```python
# Calculate average ActivityScore per user
user_avg_activity_score = (
    daily_activity_cleaned
    .groupby("Id")["ActivityScore"]
    .mean()
)

# View first few users
user_avg_activity_score.head()
```

```python
# Calculate percentile thresholds
low_threshold = user_avg_activity_score.quantile(0.33)
high_threshold = user_avg_activity_score.quantile(0.66)

# Define segmentation function
def categorize_activity_score(score):
    if score <= low_threshold:
        return "Low Activity User"
    elif score <= high_threshold:
        return "Moderate Activity User"
    else:
        return "High Activity User"

# Apply segmentation
composite_user_segments = user_avg_activity_score.apply(categorize_activity_score)

# Count users in each segment
composite_user_segments.value_counts()
```

```python
# Combine both segmentations into one dataframe
comparison_df = pd.DataFrame({
    "StepBasedSegment": user_segments,
    "CompositeSegment": composite_user_segments
})

# Create comparison table
segment_comparison = pd.crosstab(
    comparison_df["StepBasedSegment"],
    comparison_df["CompositeSegment"]
)

segment_comparison
```

```python
# Combine composite segmentation with average calories per user
composite_summary = pd.DataFrame({
    "CompositeSegment": composite_user_segments,
    "AvgCalories": user_avg_calories
})

# Calculate average calories by composite segment
composite_calorie_analysis = (
    composite_summary
    .groupby("CompositeSegment")["AvgCalories"]
    .mean()
    .sort_values(ascending=False)
)

composite_calorie_analysis
```
